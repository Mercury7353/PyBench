````python
import pygame
import random
import os

# Set up the game window
pygame.init()
win_width = 600
win_height = 600
win = pygame.display.set_mode((win_width, win_height))
pygame.display.set_caption("Snake Game")

# Set up the snake
snake_speed = 10
snake_x = win_width / 2
snake_y = win_height / 2
snake_size = 10
snake_x_change = 0
snake_y_change = 0
snake = []
for i in range(5):
    x = snake_x - (i * snake_size)
    y = snake_y
    snake.append([x, y])

# Set up the food
food_x = random.randint(50, win_width - 50)
food_y = random.randint(50, win_height - 50)

# Set up the game loop
run = True
while run:
    pygame.time.delay(50)

    # Check for events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT and snake_x_change != snake_speed:
                snake_x_change = -snake_speed
                snake_y_change = 0
            if event.key == pygame.K_RIGHT and snake_x_change != -snake_speed:
                snake_x_change = snake_speed
                snake_y_change = 0
            if event.key == pygame.K_UP and snake_y_change != snake_speed:
                snake_x_change = 0
                snake_y_change = -snake_speed
            if event.key == pygame.K_DOWN and snake_y_change != -snake_speed:
                snake_x_change = 0
                snake_y_change = snake_speed

    # Move the snake
    for i in range(len(snake) - 1, 0, -1):
        snake[i][0] = snake[i - 1][0]
        snake[i][1] = snake[i - 1][1]

    snake[0][0] += snake_x_change
    snake[0][1] += snake_y_change

    # Check for collision with food
    if snake[0][0] == food_x and snake[0][1] == food_y:
        food_x = random.randint(50, win_width - 50)
        food_y = random.randint(50, win_height - 50)
        snake.append(snake[-1])
        snake_speed += 1

    # Check for collision with walls or self
    if snake[0][0] < 0 or snake[0][0] > win_width - snake_size or snake[0][1] < 0 or snake[0][1] > win_height - snake_size:
        run = False
    for i in range(1, len(snake)):
        if snake[0][0] == snake[i][0] and snake[0][1] == snake[i][1]:
            run = False

    # Draw the game
    win.fill((0, 0, 0))
    for i in snake:
        pygame.draw.rect(win, (255, 0, 0), [i[0], i[1], snake_size, snake_size])
    pygame.draw.rect(win, (0, 255, 0), [food_x, food_y, snake_size, snake_size])

    # Update the display
    pygame.display.update()

# Save the game files
try:
    os.makedirs('./output/140/', exist_ok=True)
    pygame.image.save(win, './output/140/snake_game.png')
    with open('./output/140/snake_game.txt', 'w') as file:
        file.write("Snake game saved successfully.")
except Exception as e:
    print(f"Error saving game files: {e}")

# Quit the game
pygame.quit()
````
