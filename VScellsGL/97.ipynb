````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image from the specified path
image_path = './data/97.jpeg'
image = cv2.imread(image_path)

# Convert the image from BGR to HSV color space
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Split the HSV image into its components
h, s, v = cv2.split(hsv_image)

# Increase the saturation by a factor of 1.5
s = np.clip(s * 1.5, 0, 255)

# Merge the components back together
enhanced_image = cv2.merge([h, s, v])

# Convert the enhanced image back to BGR color space
enhanced_image_bgr = cv2.cvtColor(enhanced_image, cv2.COLOR_HSV2BGR)

# Convert the BGR image to RGB color space for displaying
enhanced_image_rgb = cv2.cvtColor(enhanced_image_bgr, cv2.COLOR_BGR2RGB)

# Display the original and enhanced images
plt.figure(figsize=(10, 5))

# Display the original image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Display the enhanced image
plt.subplot(1, 2, 2)
plt.imshow(enhanced_image_rgb)
plt.title('Enhanced Image')
plt.axis('off')

# Save the figure containing both the original and enhanced images
plt.savefig('enhanced_image.png')
plt.close()

# Print the path to the saved figure
print("Enhanced image saved as enhanced_image.png")
````
